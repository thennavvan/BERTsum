```python
!pip install transformers datasets nltk rouge-score scikit-learn --quiet

import os
import torch
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import sent_tokenize
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from sklearn.model_selection import train_test_split
from datasets import Dataset
from rouge_score import rouge_scorer

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
# Adjust path as per your setup
news_dir = "/content/BBC News Summary/News Articles"
summary_dir = "/content/BBC News Summary/Summaries"

categories = os.listdir(news_dir)
data = []

for category in categories:
    news_path = os.path.join(news_dir, category)
    summary_path = os.path.join(summary_dir, category)

    for file in os.listdir(news_path):
        with open(os.path.join(news_path, file), "r", encoding="ISO-8859-1") as nf:
            text = nf.read().strip()

        with open(os.path.join(summary_path, file), "r", encoding="ISO-8859-1") as sf:
            summary = sf.read().strip()

        data.append({"category": category, "text": text, "summary": summary})

df = pd.DataFrame(data)
```

```python
from sentence_transformers import InputExample
# Limit data for faster training (adjust as needed)
df = df[:2000]

# Create SBERT input examples with a label (1.0 means highly similar)
train_examples = [
    InputExample(texts=[row["text"], row["summary"]], label=1.0)
    for _, row in df.iterrows()
]

```

```python
from sentence_transformers import SentenceTransformer, losses
from torch.utils.data import DataLoader

# Load SBERT
model = SentenceTransformer("paraphrase-MiniLM-L6-v2")

# DataLoader
train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=16)

# Loss function: CosineSimilarityLoss for sentence pairs
train_loss = losses.CosineSimilarityLoss(model=model)
```

```python
model.fit(
    train_objectives=[(train_dataloader, train_loss)],
    epochs=2,  # You can increase to 2–3 for better results
    warmup_steps=100,
    output_path="./fine_tuned_sbert"
)
```

```python
from nltk.tokenize import sent_tokenize
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load fine-tuned model
model = SentenceTransformer("./fine_tuned_sbert")

def extract_summary(text, model, top_k=3):
    sentences = sent_tokenize(text)
    sentence_embeddings = model.encode(sentences)
    doc_embedding = model.encode([text])[0]

    similarities = cosine_similarity([doc_embedding], sentence_embeddings)[0]
    top_indices = similarities.argsort()[-top_k:]
    top_sentences = [sentences[i] for i in sorted(top_indices)]

    return " ".join(top_sentences)

# Example
sample_article = df.iloc[0]["text"]
print("Generated Summary:\n", extract_summary(sample_article, model))
```
